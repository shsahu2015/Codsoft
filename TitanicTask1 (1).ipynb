```python
q25=df["Age"].quantile(0.25)
q75=df["Age"].quantile(0.75)
iqr=q75-q25
ul=q75+1.5*iqr
ll=q25-1.5*iqr
df['Age']=np.where(df['Age']>ul,ul,np.where(df['Age']<ll,ll,df['Age']))
```

```python
q25=df["SibSp"].quantile(0.25)
q75=df["SibSp"].quantile(0.75)
iqr=q75-q25
ul=q75+1.5*iqr
ll=q25-1.5*iqr
df['SibSp']=np.where(df['SibSp']>ul,ul,np.where(df['SibSp']<ll,ll,df['SibSp']))
```

```python
plt.figure(figsize=(12,12))
plt.suptitle("CHECKING OUTLIERS")
pltn=1
for i in numcol:
 if pltn<=6:
  ax=plt.subplot(4,2,pltn)
  sns.boxplot(df[i])
  plt.xlabel(i,fontsize=15)
  pltn=pltn+1
plt.tight_layout()
```
